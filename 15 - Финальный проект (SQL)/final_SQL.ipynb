## Анализ данных

**Посчитаем, сколько книг вышло после 1 января 2000 года**

```python
query_count_books = ''' 
SELECT COUNT(book_id)
FROM books
WHERE publication_date > '2000-01-01';
'''

query(query_count_books)
```

***Вывод:*** 

После 1 января 2000 года вышло 819 книг

**Для каждой книги посчитаем количество обзоров и среднюю оценку**

```python
query_count_books_1 = ''' 
SELECT b.book_id, b.title, COUNT(DISTINCT rev.review_id) AS count_review, AVG(rat.rating) AS mean_rating
FROM books AS b 
LEFT JOIN reviews AS rev ON b.book_id = rev.book_id
LEFT JOIN ratings AS rat ON rev.book_id = rat.book_id
GROUP BY b.book_id
ORDER BY count_review DESC
LIMIT 10;
'''

query(query_count_books_1)
```

```python
query_count_books_2 = ''' 
SELECT b.book_id, b.title, COUNT(DISTINCT rev.review_id) AS count_review, AVG(rat.rating) AS mean_rating
FROM books AS b 
LEFT JOIN reviews AS rev ON b.book_id = rev.book_id
LEFT JOIN ratings AS rat ON rev.book_id = rat.book_id
GROUP BY b.book_id
ORDER BY mean_rating DESC;
'''

query(query_count_books_2)
```

***Вывод:***

В нашей базе максимальное количество обзоров (7) пришлось на книгу "Twilight" ("Сумерки"). 
Есть книги, по которым не было написано ни одного обзора и нет оценок.

**Определим издательство, которое выпустило наибольшее число книг толще 50 страниц — так мы исключим из анализа брошюры**

```python
query_publisher = ''' 
SELECT p.publisher, p.count_books
FROM (SELECT b.publisher_id AS publisher_id,
      COUNT(b.book_id) AS count_books,
      publishers.publisher AS publisher 
      FROM books AS b
      LEFT JOIN publishers ON b.publisher_id=publishers.publisher_id
      WHERE num_pages>50
      GROUP BY b.publisher_id, publishers.publisher
      ORDER BY count_books DESC) AS p
LIMIT 1;
'''

query(query_publisher)
```

***Вывод:*** 

Мы нашли издательство, которое выпустило наибольшее число книг толще 50 страниц. Это издательство называется "Penguin Books"

**Определим автора с самой высокой средней оценкой книг (учитывать будем только книги с 50 и более оценками)**

```python
query_author = ''' 
SELECT a.author, AVG(r.rating) AS mean_rating
FROM books AS b 
JOIN ratings AS r ON b.book_id = r.book_id
JOIN authors AS a ON b.author_id = a.author_id
WHERE b.book_id IN (
                     SELECT book_id
                     FROM ratings
                     GROUP BY book_id
                     HAVING COUNT(rating_id) >= 50
                    )
GROUP BY a.author
ORDER BY mean_rating DESC
LIMIT 1;

'''

query(query_author)
```

***Вывод:*** 

Автором с самой высокой средней оценкой книг является Джоан Роулинг (J.K. Rowling/Mary GrandPré), написавшая романы о Гарри Поттере

**Посчитаем среднее количество обзоров от пользователей, которые поставили больше 50 оценок**

```python
query_mean_reviews = ''' 
SELECT AVG(count_review) AS mean_count_review
FROM (
        SELECT rev.username, COUNT(DISTINCT rev.review_id) AS count_review
        FROM reviews AS rev
        WHERE rev.username IN (SELECT username
                               FROM ratings
                               GROUP BY username
                               HAVING COUNT(DISTINCT rating_id) > 50)
                               GROUP BY username) AS u 

'''

query(query_mean_reviews)
```

***Вывод:*** 

Активные пользователи, которые поставили более 50 оценок, в среднем написали по 24 обзора.
